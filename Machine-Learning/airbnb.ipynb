Checking Outliers

```python
import matplotlib.pyplot as mlt
```

```python
for i in df_new:
    if df_new[i].dtype=="int" or df_new[i].dtype=="float":
        mlt.title(f"Boxplot of {i}")
        mlt.boxplot(df_new[i])
        mlt.show()
```

Removing Outliers

```python
li = ["availability 365","minimum nights"]
outlier_dict = {}
for col in li:
    q1 = df_new[col].quantile(0.25)
    q3 = df_new[col].quantile(0.75)
    iqr = q3-q1
    lower = q1-1.5*iqr
    upper = q3+1.5*iqr
    outlier_dict[col]=(lower,upper)
print(outlier_dict)
```

```python
for col in outlier_dict:
    mn,mx = outlier_dict[col]
    df_new = df_new.loc[(df_new[col]>=mn)&(df_new[col]<=mx)]
```

```python
for i in df_new:
    if df_new[i].dtype=="int" or df_new[i].dtype=="float":
        mlt.title(f"Boxplot of {i}")
        mlt.boxplot(df_new[i])
        mlt.show()
```
